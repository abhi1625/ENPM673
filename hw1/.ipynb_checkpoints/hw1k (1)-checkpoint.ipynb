```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

with open('/home/kamakshi/Desktop/hw1/data1_py2.pkl', 'rb') as f:
     data = pickle.load(f)
        
X = np.asarray(data)
cov_matrix = np.cov(X.T) #transposed to convert data from nx2 to 2xn matrix
print("The Covariance Matrix is = \n", cov_matrix)
plt.scatter(X[:,0],X[:,1]) # since not transposed its a nx2 matrix
#plt.show()

mean = X-np.mean(X,axis=0)  #axis=0 is mean for individual rows
print("Mean = ", mean)

eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print("Eigen Value=",eig_vals)
print("Eigen Vectors \n",eig_vecs)
origin = mean
x = eig_vecs[0,:] 
y = eig_vecs[1,:] 
#print(x,y)
plt.quiver(*origin, x, y, color=['r','b'], scale=1)
plt.show()
```

```python
#scaling matrix
sx= 0.5
sy=0.4
Scale = np.array([[sx, 0], [0, sy]])

#applying matrix
Scaled_X = X.dot(Scale)
plt.scatter(Scaled_X [:, 0], Scaled_X [:, 1],label='old data')
Scaled_Cov =  np.cov(Scaled_X.T)
print("Scaled_Cov \n",Scaled_Cov)

#rotation matrix
theta = 0.8*np.pi
cos, sin = np.cos(theta), np.sin(theta)
Rotation = np.array([[cos, -sin], [sin, cos]])

# Transformation matrix
T = Scale.dot(Rotation)
print("Transformation matrix after Rotation n scaling \n",T)

# Apply transformation matrix to X
D = X.dot(T)
#print("data after transformation \n", D)
plt.scatter(D[:, 0], D[:, 1],label='transformed data')
plt.legend()
plt.show()





```

```python
C_Mt= np.cov(D.T)
eig_Vat, eig_Vect = np.linalg.eig(C_Mt)

plt.scatter(D[:, 0], D[:, 1])
for s, r in zip(eig_Vat, eig_Vect.T):
    plt.plot([0, 3*np.sqrt(s)*r[0]], [0, 3*np.sqrt(s)*r[1]], 'r-', lw=3)


```
